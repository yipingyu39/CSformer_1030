```python
import re
import  matplotlib.pyplot as plt
```

```python
log = 'logs/training_log_20240601_222527_pool_ln_data_coordinate.log'
```

```python
with open(log) as f:
    lines = f.readlines()
loss_list = []

for line in lines:
    if 'BinaryCrossEntropyBackward' in line:
        match = re.search(r'tensor\(([\d\.]+)', line)
        loss_value = float(match.group(1))
        loss_list.append(loss_value)
```

```python
plt.plot(loss_list)
```

```python
train_loss = []
for line in lines:
    if 'train' in line:
        auc = line.split(':')[-1]
        train_loss.append(float(auc))
plt.plot(train_loss)
```

```python
train_loss
```

```python
test_loss = []
for line in lines:
    if 'test' in line:
        auc = line.split(':')[-1]
        test_loss.append(float(auc))
plt.plot(test_loss)
```
