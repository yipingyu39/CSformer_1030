```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def draw_node_feautures(data,figname):
    index_list = [199,221,178,190,232,243,257,263]
    y_data = data
    plt.figure(figsize=(10, 8))
    plt.grid(None)
    
    # 定义 x 轴和 y 轴的刻度位置
    x_ticks = np.arange(0, y_data.shape[0], 1)  # 例如，每 10 个单位一个刻度
    x_data = [x+25 for x in x_ticks]
    plt.plot(x_data,y_data)
    # 定义刻度标签
    x_tick_labels = [f"res{tick+25}" for tick in x_ticks]  # 例如，标签为 'x0', 'x50', ...
    #plt.xlim(x_ticks[0]+30,x_ticks[-1]+30)
    # 设置刻度位置和标签
    #plt.xticks(x_ticks, x_tick_labels,rotation=90)
    plt.xlabel('Residue',fontproperties = 'Times New Roman',fontsize=24)
    #plt.title("Node Feature Gradients",fontproperties = 'Times New Roman',fontsize=24)
    plt.xticks(fontproperties = 'Times New Roman',fontsize=20)
    plt.yticks(fontproperties = 'Times New Roman',fontsize=20)
    plt.ylim(0,1.6)
    plt.savefig(figname,dpi=600)
```

```python
def draw_graph(df,output):
    index_list = [25,32,40,45,64,91,98,113,152,164,172,177,199,221,178,190,232,243,257,263,271,282,296,310,319,352]
    index_list = [index-25 for index in index_list]
    # 可视化边特征梯度
    # 使用平方根规范化

    
    plt.figure(figsize=(10, 8))
    plt.grid(None)
    plt.imshow(df, cmap='jet', interpolation='nearest')  # 取平均值作为边的代表
    
    # 定义 x 轴和 y 轴的刻度位置
    x_ticks = np.arange(0, df.shape[1], 20)  # 例如，每 50 个单位一个刻度
    y_ticks = np.arange(0, df.shape[0], 20)
    #ticks = np.array([3.5,11,17.5,26,52.5])
    #tick_labels = ['αN','β1','P-loop','α1','α2']
    
    # 定义刻度标签
    x_tick_labels = [f"{tick+25}" for tick in x_ticks]  # 例如，标签为 'x0', 'x50', ...
    y_tick_labels = [f"{tick+25}" for tick in y_ticks]  # 例如，标签为 'y0', 'y50', ...
    # 设置刻度位置和标签
    plt.xticks(x_ticks, x_tick_labels,rotation=90,fontproperties = 'Times New Roman',fontsize=20)
    plt.yticks(y_ticks, y_tick_labels,fontproperties = 'Times New Roman',fontsize=20)
    plt.xlabel('Residue',fontsize=24,fontproperties = 'Times New Roman')
    plt.ylabel('Residue',fontsize=24,fontproperties = 'Times New Roman')
    
    #plt.title("Edge Feature Gradients",fontproperties = 'Times New Roman',fontsize=24)
    plt.colorbar(pad=0.1)
    plt.savefig(output,dpi=600)
```

```python
filename  = 'input/superpose/superpose_GaoA-GDPG.npz'
data = np.load(filename)['arr_0']
data = data.squeeze()
avg_node = np.mean(data,axis=0)
draw_node_feautures(avg_node,'node_feature_distribution-GaoA-GDPG.png')
```
