```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def get_variance(system,time_scale):
    #system = 'GaoB-GTPG-MG'
    filenames = os.listdir(f'input/data/{system}')
    rp_list = []
    for filename in filenames:
        data = np.load(os.path.join(f'input/data/{system}',filename),allow_pickle=True).item()
        rp_list.append(data[time_scale])
    # 将列表转换为一个三维的NumPy数组
    stacked_data = np.stack(rp_list, axis=0)
    # 计算每个位置的方差
    variance = np.var(stacked_data, axis=0)
    plt.imshow(variance)
    plt.colorbar()
    return variance
```

```python
get_variance('GaoB-GTPG-MG','rp_1ns')
```
